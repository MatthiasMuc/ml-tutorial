### Aufgabe 2. Polynomiale Regression

1. Implementieren Sie polinomiale Modelle (grad von 1 bis 7) für die Regressionsaufgabe. Visualisieren Sie alle Modelle auf einem Graphen. Vergleichen Sie auch den Fehler. 

```python
print(studying)
# preparation for fitting: transform vector to matrix
# fitting needs a matrix:
# - rows are the samples
# - columns are the features
# in this case: 14 rows, 1 col
studyHrsArr = np.array(studying)
# 1 col, as many rows as needed
studyHrsArr = studyHrsArr.reshape(-1, 1)
# print(studyHrsArr)

# prepare visualization:
# fig = plt.figure(figsize=(15, 15))
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot()
ax.scatter(studying, result)
ax.set_title("Polynomiale Regression")
ax.set_xlabel("Vorbereitungszeit auf Klausur in Minuten")
ax.set_ylabel("Klausurergebnis")


model = LinearRegression()
mse = []

for i in range(1, 8):
    poly = PolynomialFeatures(i)
    # fitting for input data: calculate all data relevant for polynom calculations, i.e.:
    # for each value (student learning time here):
    #    - x**i, up to polynom degree (x**0, x**1, x**2, ...) (for multiple features done for each feature)
    #    - multiplication of features: x*y
    studyingPoly = poly.fit_transform(studyHrsArr)
    print("studyingPoly:\n", studyingPoly)

    # Trainiere
    model.fit(studyingPoly, result)

    # predict values
    prediction = model.predict(studyingPoly)
    print("predict:\n", prediction)

    # extend prediction for printing
    xAxis = np.linspace(0, 1400).reshape(-1, 1)
    predictionPrint = model.predict(poly.fit_transform(xAxis))
    print("predict for print:\n", predictionPrint)

    # calculate error
    mseTmp = mseCalc(prediction, result)
    print("mseTmp: ", mseTmp)
    mse.append(mseTmp)

    # visualize
    label = "Ord: {}, MSE: {}".format(i, round(mse[i - 1], 2),)
    ax.plot(xAxis, predictionPrint, label=label)

    print(
        "Polynom order {}: MSE: {}, bias: {}".format(
            str(i), mse[i - 1], model.intercept_
        )
    )

print("MSE:\n", mse)
ax.legend()
```
