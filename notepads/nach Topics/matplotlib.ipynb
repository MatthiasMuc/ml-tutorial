```python
# create a figure object for plots
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
#ax4 = fig.add_subplot(2, 2, 4)

# plot some values (default: last subplot is used)
ax3.plot([4, 6, 8, 27, 33, 88, 999])

# plot a histogram with 10 bins for 100 values
npArr = np.random.randn(100)
ax1.hist(npArr)
#ax2.hist(npArr, alpha=0.3) # alpha = 0.3 -> farbe ist heller
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
```

```python
# General process:
fig, axes = plt.subplots(2, 2)  # further params: sharex, sharey

axes[0, 0].hist(npArr)
axes[0, 1].scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
axes[1, 0].plot([4, 6, 8, 27, 33, 88, 999],linestyle='dashed', marker='o')
ax = axes[1,0]
plt.subplots_adjust(hspace=0.4) #provide enough space for title
ax.set_title("My first plot in mpl")
#ax1.plot?
#axes[0,0].plot?
```
